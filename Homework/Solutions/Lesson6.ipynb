### Object-Oriented Programming - Homework

```python
# Dog is an animal with four legs, can make sound "Bow Bow" and can run.
# Peacock is a bird with two legs, two wings and can make sound "Eeeiu" and can fly.
# Bird is an Animal category

# List all the objects possible from above statements with their attributes and behaviours.
#Start Here

## Dog is an object with legs as attribute and makeSound, run as behaviours/methods
## Peacock is an object with legs, wings as attributes and makeSound, fly as behaviours/methods


#Define classes for objects listed above
#Start Here

class Dog:                   
    'This class represent animal Dog'
    def __init__(self, no_of_legs):
        self.legs = no_of_legs
    def makeSound(self):
        print('Bow Bow')
    def run(self):
        print('Runing')

class Peacock:
    'This class represent Bird peacock'
    def __init__(self, no_of_legs, no_of_wings):
        self.legs = no_of_legs
        self.wings = no_of_wings
    
    def makeSound(self):
        print('Eeeiu')
    
    def fly(self):
        print('Flying')

#Draw the relations between the classes defined above with common attributes and behaviours
#Start Here

##Common attributes is legs
##Common behaviours is makeSound

#Define parent classes with common behaviours and inherit thoose with original classes
#Start Here

#Create a dog with name "Cookie" and peacock with name "Jewel"
#Start Here
class Animal:                   
    'This class represent animal'
    def __init__(self, no_of_legs, sound):
        self.legs = no_of_legs
        self.sound = sound

    def makeSound(self):
        return self.sound

class Dog(Animal):
    'This class represent animal Dog'
    def __init__(self, name, no_of_legs):
        Animal.__init__(self,no_of_legs, "Bow Bow")
        self.name = name   
        
    def printDetails(self):
        print("Dog name:", self.name, "\n Sound:", super(Dog, self).makeSound())

class Bird(Animal):                   
    'This class represent Bird'
    def __init__(self, no_of_legs, no_of_wings, sound):
        Animal.__init__(self, no_of_legs, sound)
        self.wings = no_of_wings
        
    def fly(self):
        return 'Flying'

class Peacock(Bird):
    'This class represent Bird peacock'
    def __init__(self, name, no_of_legs, no_of_wings):
        Bird.__init__(self, no_of_legs, no_of_wings, "Eeeiu")
        self.name = name
    def printDetails(self):
        print("Peacock name:", self.name, "\n Sound:", super(Bird, self).makeSound(), "\n Can perform:", self.fly())
        
#Create a dog with name "Cookie" and peacock with name "Jewel"
#Start Here

cookie = Dog("Cookie", 4);
cookie.printDetails()
jewel = Peacock("Jewel", 2, 2);
jewel.printDetails()

##Output will be as below
#Dog name: Cookie 
# Sound: Bow Bow
#Peacock name: Jewel 
# Sound: Eeeiu 
# Can perform: Flying
```

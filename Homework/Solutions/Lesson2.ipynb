#### Exercise 1:
1. Ask the user for his firstname
2. Ask the user for his lastname
3. Ask the user for his gender
4. Ask the user if he's married
5. Print the following message "Hello {title} {name}, your initials are: {initials}"
    * Where {title} is either Mr., Mrs., or Miss. depending on the user's answers
    * Where {initials} is the first letter of his firstname and lastname.

```python
# 1. Ask the user for his firstname
# 2. Ask the user for his lastname
# 3. Ask the user for his gender
# 4. Ask the user if he's married
# 5. Print the following message "Hello {title} {name}, your initials are: {initials}"
#    - Where {title} is either Mr., Mrs., or Miss. depending on the user's answers
#    - Where {initials} is the first letter of his firstname and lastname.

fn = input("What is your First name? ")
ln = input("What is your Last name? ")
gender = input("What is your gender? ")
married = input("Are you married? ") == "yes"

if gender=="female" and married:
    title = "Mrs."
elif gender=="female":
    title = "Miss."
else:
    title = "Mr."
    
print("Hello", title, fn, ln, "your initials are:", fn[0]+ln[0])
```

#### Exercise 2:
1. Ask the user for his age
2. Ask the user for the list of all his siblings age in order
3. Print the number of siblings
4. Print the birth order of the user among his siblings

```python
# 1. Ask the user for his age
# 2. Ask the user for the list of all his siblings age in order
# 3. Print the number of siblings
# 4. Print the birth order of the user among his siblings

age = input("How old are you? ")
siblingsAge = input("How old are your siblings? ")

sbiList = siblingsAge.split(" ")

if siblingsAge=="":
    print("You are an only child!")
else:
    print("You have", len(sbiList) , "siblings")
    if age < sbiList[0]:
        order = "last"
    elif age > sbiList[len(sbiList)-1]:
        order = "1st"
    elif age < sbiList[1]:
        order = "2nd"
    elif age < sbiList[2]:
        order = "3rd"
    elif age < sbiList[3]:
        order = "4th"
    elif age < sbiList[4]:
        order = "5th"
    elif age < sbiList[5]:
        order = "6th"
    elif age < sbiList[6]:
        order = "7th"
    
    if not order == None:
        print("You are the", order , "sibling")
```

#### Exercise 3:
1. Imagine a team of 3 persons. Write every team member's details (name, age, address, hobbies) in a variable.
2. Ask the user for the number of the team member he wants to see the details of and print it
3. Ask the user for a team member's name and print his details.

```python
# 1. Imagine a team of 3 persons. Write every team member's details (name, age, address, hobbies) in a variable.
team_members = [
    {
        'name': 'Alice',
        'age': 12,
        'hobbies': ['reading', 'singing'],
        'address': {
            'city': 'Paris',
            'country': "France",
            'street': 'Rue de la Paix',
            "number": 10
        }
    },
    {
        'name': 'Bob',
        'age': 21,
        'hobbies': ['dancing', 'drawing'],
        'address': {
            'city': 'Singapore',
            'country': "Singapore",
            'street': 'Mackenzie road',
            "number": 33
        }
    },
    {
        'name': 'Jason',
        'age': 24,
        'hobbies': ['CSR', 'workout'],
        'address': {
            'city': 'Singapore',
            'country': "Singapore",
            'street': 'Beach road',
            "number": 36
        }
    }
]


# 2. Ask the user for the number of the team member he wants to see the details of and print it
memberIndexString = input("Which team member do you want to see the details of (use index) ? ")
memberIndex = int(memberIndexString)
if memberIndex >= 0 and memberIndex < 3:
    print("Here's team member #"+memberIndexString+" details:", team_members[memberIndex])
else:
    print("Nothing to show!")

# 3. Ask the user for a team member's name and print his details.
memberName = input("Which team member do you want to see the details of (use name) ? ")

if team_members[0]["name"] == memberName:
    details = team_members[0]
elif team_members[1]["name"] == memberName:
    details = team_members[1]
elif team_members[2]["name"] == memberName:
    details = team_members[2]
else:
    details = "Nothing to show!"

print("Here's team member '"+memberName+"' details:", details)
```
